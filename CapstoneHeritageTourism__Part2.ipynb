## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt

# Define file paths
tourism_rating_path = "/content/drive/MyDrive/tourism_rating.csv"
tourism_with_id_path = "/content/drive/MyDrive/tourism_with_id.xlsx"
user_path = "/content/drive/MyDrive/user.csv"

# Load the datasets
tourism_rating_df = pd.read_csv(tourism_rating_path)
tourism_with_id_df = pd.read_excel(tourism_with_id_path)
user_df = pd.read_csv(user_path)

# Merge tourism_rating_df with user_df on User_Id to get user demographics for ratings
tourism_user_df = tourism_rating_df.merge(user_df, on="User_Id", how="left")

# Analyze Age Distribution
plt.figure(figsize=(10, 5))
plt.hist(tourism_user_df["Age"], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.title("Age Distribution of Tourists Rating Places")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Analyze Tourist Locations (Top 10 locations where users come from)
top_locations = tourism_user_df["Location"].value_counts().head(10)

# Plot the top 10 user locations
plt.figure(figsize=(12, 5))
top_locations.plot(kind="bar", color="royalblue", edgecolor='black', alpha=0.7)
plt.xlabel("Location")
plt.ylabel("Number of Tourists")
plt.title("Top 10 Locations Where Tourists Come From")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Display the top locations in a structured format
print ("Top Tourist Locations", top_locations.to_frame())
```

```
# Re-load the datasets after execution state reset
import pandas as pd

# 1. Identify different categories of tourist spots
unique_categories = tourism_with_id_df["Category"].value_counts()

# 2. Identify what kind of tourism each location is most famous for
category_by_city = tourism_with_id_df.groupby("City")["Category"].value_counts().unstack()

# 3. Identify the best city for a nature enthusiast (assuming 'Alam' represents nature-based tourism)
best_nature_city = tourism_with_id_df[tourism_with_id_df["Category"] == "Alam"]["City"].value_counts()

# Display results

print("Tourist Spot Categories",unique_categories.to_frame())
print("Tourism by City",category_by_city)
print("Best City for Nature Enthusiasts",best_nature_city.to_frame())
```

```
# Merge tourism ratings with place details
combined_df = tourism_rating_df.merge(tourism_with_id_df, on="Place_Id", how="left")

# 1. Identify the most loved tourist spots (based on average ratings)
most_loved_spots = combined_df.groupby(["Place_Name", "City"])["Place_Ratings"].mean().sort_values(ascending=False)

# 2. Identify the city with the most highly rated tourist spots
most_loved_city = combined_df.groupby("City")["Place_Ratings"].mean().sort_values(ascending=False)

# 3. Identify the most liked category of places
most_liked_category = combined_df.groupby("Category")["Place_Ratings"].mean().sort_values(ascending=False)

# Display results

print("Most Loved Tourist Spots", most_loved_spots.to_frame())
print("Most Loved Cities", most_loved_city.to_frame())
print("Most Liked Categories", most_liked_category.to_frame())
```

```
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Reload datasets
tourism_rating_df = pd.read_csv("/content/drive/MyDrive/tourism_rating.csv")
tourism_with_id_df = pd.read_excel("/content/drive/MyDrive/tourism_with_id.xlsx")

# Create a user-item rating matrix
rating_matrix = tourism_rating_df.pivot_table(index="User_Id", columns="Place_Id", values="Place_Ratings", aggfunc='mean')

# Fill missing values with 0 (implicit feedback method)
rating_matrix.fillna(0, inplace=True)

# Compute cosine similarity between users
user_similarity = cosine_similarity(rating_matrix)

# Function to get place recommendations for a given user using user-based collaborative filtering
def get_recommendations_cf(user_id, num_recommendations=5):
    if user_id not in rating_matrix.index:
        return pd.DataFrame()  # Return empty if user not in dataset

    # Find similar users
    user_idx = rating_matrix.index.get_loc(user_id)
    similar_users = np.argsort(-user_similarity[user_idx])[1:6]  # Top similar users (excluding self)

    # Get places rated by similar users but not yet rated by the target user
    similar_users_ratings = rating_matrix.iloc[similar_users].mean(axis=0)
    user_rated_places = rating_matrix.loc[user_id][rating_matrix.loc[user_id] > 0].index
    recommendations = similar_users_ratings.drop(user_rated_places).sort_values(ascending=False).head(num_recommendations)

    # Fetch recommended place details
    recommended_places = tourism_with_id_df[tourism_with_id_df["Place_Id"].isin(recommendations.index)]

    return recommended_places

# Example: Generate recommendations for user with ID 1
recommendations_df_cf = get_recommendations_cf(user_id=1, num_recommendations=5)

# Display recommendations

print("Recommended Tourist Spots", recommendations_df_cf)
```
